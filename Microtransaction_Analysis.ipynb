```python
# Set up pandas module
import pandas as pd

# load data file
file = "Resources/purchase_data.csv"

# Read microtransaction csv and display as dataframe
purchase_data = pd.read_csv(file)
purchase_data = pd.DataFrame(purchase_data)
purchase_data.head()
```

## Player Count

```python
# calculate the total number of unique players
TotalPlayers = purchase_data['SN'].nunique()

print(f'Total Players: {TotalPlayers}')
```

## Purchasing Analysis (Total)

```python
# calculate the total number of unique items, average item price, total purchaes, and total revenue
totaluniqueitems = purchase_data['Item Name'].nunique()
avgprice = round(purchase_data['Price'].mean(), 2)
numpurchases = purchase_data['Purchase ID'].count()
totalrevenue = purchase_data['Price'].sum()

# create a list of values and headers. Create a dataframe
values = [[totaluniqueitems,avgprice,numpurchases,totalrevenue]]
headers = ['Number of Unique Items','Average Price','Number of Purchases','Total Revenue']
table = pd.DataFrame(values, columns = headers)

#format table with proper units and display
table['Average Price'] = table['Average Price'].map("${}".format)
table['Total Revenue'] = table['Total Revenue'].map("${}".format)
table
```

## Gender Demographics

```python
# create a subset of the microtransaction data
gender_data1 = purchase_data[['Gender','SN']].copy()

# group data by gender
gender_group = gender_data1.groupby(['Gender'])

# calculate player count by gender, create a dataframe, and rename first column
gender_count = gender_group['SN'].nunique()
gender_percent = round(gender_group['SN'].nunique()/TotalPlayers * 100, 1)

# calculate the percentage of total players by gender. Create a Percent column within the previous dataframe
gender_merge = {'Player Count': gender_count,
            'Percent of Total Players': gender_percent}
gender_merge = pd.DataFrame(gender_merge)

# sort values based on descending player counts, reformat percentages, and display
gender_merge = gender_merge.sort_values('Player Count', ascending=False)
gender_merge['Percent of Total Players'] = gender_merge['Percent of Total Players'].map("{}%".format)
gender_merge
```

## Age Demographics

```python
# create bins and bin labels for age groups
bins = [0, 9, 14, 19, 24, 29, 34, 39, 44, 49]
group_names = ['<10', '10-14', '15-19', '20-24','25-29', '30-34', '35-39','40-44','45-49']

# create a subset of microtransaction data, create age groups, and group by age group
age_data = purchase_data[['Age','Purchase ID','SN','Price']].copy()
age_data['Age Group'] = pd.cut(age_data['Age'], bins, labels = group_names, include_lowest=True)
age_group = age_data.groupby(['Age Group'])

# calculate the player count and percent of total players by age group. Create dataframe
age_count = age_group['SN'].nunique()
age_count_df = {'Player Count':age_count}
age_count_df = pd.DataFrame(age_count_df)

# format table properly and display
age_count_df['Percent of Total Players'] = round(age_count_df['Player Count']/TotalPlayers * 100, 1)
age_count_df['Percent of Total Players'] = age_count_df['Percent of Total Players'].map("{}%".format)
age_count_df
```

## Purchasing Analysis (Age)

```python
# calculate the number of purchases, avg purchase, total purchase price, and avg purchase per person by age group.
age_npurch = age_group['Purchase ID'].count()
age_avgprice = round(age_group['Price'].mean(), 2)
age_totalprice = round(age_group['Price'].sum(), 2)
age_avgperperson = round(age_totalprice/age_count, 2)

# create age dictionary and dataframe
age_merge = {'Purchase Count':age_npurch,
            'Avg. Purchase Price': age_avgprice,
            'Total Purchase Price': age_totalprice,
            'Avg Pur. Total Per Person': age_avgperperson}
age_merge = pd.DataFrame(age_merge)

# format age demographics table properly and display
age_merge['Avg. Purchase Price'] = age_merge['Avg. Purchase Price'].map("${}".format)
age_merge['Total Purchase Price'] = age_merge['Total Purchase Price'].map("${}".format)
age_merge['Avg Pur. Total Per Person'] = age_merge['Avg Pur. Total Per Person'].map("${}".format)
age_merge
```

## Top Spenders

```python
# create a subset of the microtransaction data and group by spenders
topspend_data = purchase_data[['SN','Purchase ID','Price']].copy()
topspend_group = topspend_data.groupby(['SN'])

# calculate number of purchases, avg purchase amount, total purchase amount by spender 
topspend_npurch = topspend_group['Purchase ID'].count()
topspend_avgprice = round(topspend_group['Price'].mean(), 2)
topspend_totalprice = round(topspend_group['Price'].sum(), 2)

# create a dictionary and dataframe of the values
topspend_merge = {'Purchase Count':topspend_npurch,
                'Avg. Purchase Price': topspend_avgprice,
                'Total Purchase Price': topspend_totalprice}
topspend_merge = pd.DataFrame(topspend_merge)

# sort values based on total spending, format the table, and display
topspend_merge = topspend_merge.sort_values('Total Purchase Price', ascending=False)
topspend_merge['Avg. Purchase Price'] = topspend_merge['Avg. Purchase Price'].map("${}".format)
topspend_merge['Total Purchase Price'] = topspend_merge['Total Purchase Price'].map("${}".format)
topspend_merge.head()
```

## Most Popular Items

```python
# create a subset of the microtransaction data and group by item
item_data = purchase_data[['Item ID','Item Name','Price', 'Purchase ID']].copy()
item_group = item_data.groupby(['Item ID'])

# calculate item names, number purchased, avg price, and total price by item 
item_name = item_group['Item Name'].first()
item_npurch = item_group['Purchase ID'].count()
item_avgprice = round(item_group['Price'].mean(), 2)
item_totalprice = round(item_group['Price'].sum(), 2)

# create a dictionary and dataframe of the values
item_merge = {'Item Name': item_name,
                'Purchase Count':item_npurch,
                'Avg. Item Price': item_avgprice,
                'Total Purchase Price': item_totalprice}
item_merge = pd.DataFrame(item_merge)

# sort values based on number of purchases, format the table, and display
item_merge = item_merge.sort_values('Purchase Count', ascending=False)
item_merge['Avg. Item Price'] = item_merge['Avg. Item Price'].map("${}".format)
item_merge.head()
```

## Most Profitable Items

```python
# sort values based on total purchase price, format the table, and display
item_merge = item_merge.sort_values('Total Purchase Price', ascending=False)
item_merge['Total Purchase Price'] = item_merge['Total Purchase Price'].map("${}".format)
item_merge.head()
```
